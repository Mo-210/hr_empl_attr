```python
# Importing necessary libraries for data analysis and visualization
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import plotly.graph_objs as go
import plotly.express as px
plt.style.use('Solarize_Light2')
plt.style.context('grayscale')
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
# Reading the Excel file 
# Ensure the file path is correct and the file exists in the specified location
df = pd.read_excel("hr_empl_attr.xlsx")                 
df.head()
```

```python
# Filter only employees who have left (attrition = 'Yes')
df_attrited = df[df['Attrition'] == 'Yes']

# Bin Total Working Years into ranges
bins = [0, 2, 5, 10, 15, 20, 30, 40]
labels = ['0-2', '3-5', '6-10', '11-15', '16-20', '21-30', '30+']
df_attrited['Experience Group'] = pd.cut(df_attrited['Total Working Years'], bins=bins, labels=labels, right=False)

# Count attrition per group
attrition_counts = df_attrited['Experience Group'].value_counts().sort_index()

# Plot the bar chart
plt.figure(figsize=(10,6))
sns.barplot(x=attrition_counts.index, y=attrition_counts.values, palette="Blues_d")

plt.title('Attrition Count by Years of Experience', fontsize=14)
plt.xlabel('Years of Experience (Grouped)', fontsize=12)
plt.ylabel('Number of Attritions', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Filter only attrited employees
df_attrited = df[df['Attrition'] == 'Yes']

# Count attrition by each year of experience
attrition_by_year = df_attrited['Total Working Years'].value_counts().sort_index()

# Plot
plt.figure(figsize=(12,6))
sns.barplot(x=attrition_by_year.index, y=attrition_by_year.values, color='skyblue')

plt.title('Attrition Count by Total Working Years (Individual)', fontsize=14)
plt.xlabel('Total Working Years (Experience)', fontsize=12)
plt.ylabel('Number of Attritions', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

The bar chart reveals that employee attrition is most pronounced in the early stages of a career, particularly after one year of experience, and also shows notable spikes around 5-7 and 10 years of total working experience. Conversely, attrition significantly decreases as employees gain more experience, becoming very low for those with 25 or more years, suggesting a challenge in retaining newer and mid-career individuals compared to long-tenured employees.

```python
# Filter attrited employees only
df_attrited = df[df['Attrition'] == 'Yes']

# Count attritions by job role
job_role_counts = df_attrited['Job Role'].value_counts()

# Calculate percentages
total_attritions = len(df_attrited)
job_role_percent = (job_role_counts / total_attritions * 100).round(2)

# Combine into a dataframe for plotting
job_attrition_df = pd.DataFrame({
    'Job Role': job_role_counts.index,
    'Attrition Count': job_role_counts.values,
    'Attrition %': job_role_percent.values
})

# Sort for cleaner visuals
job_attrition_df = job_attrition_df.sort_values(by='Attrition Count', ascending=False)

# Plot
plt.figure(figsize=(12,7))
barplot = sns.barplot(
    x='Attrition Count',
    y='Job Role',
    data=job_attrition_df,
    palette='mako'
)

# Annotate bars with percentage labels
for index, row in job_attrition_df.iterrows():
    barplot.text(
        row['Attrition Count'] + 0.5,
        index,
        f"{row['Attrition %']}%",
        color='black',
        va='center',
        fontsize=10
    )

plt.title("Attrition Count by Job Role with Percentage of Total Attrition", fontsize=14)
plt.xlabel("Number of Attritions")
plt.ylabel("Job Role")
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

The provided bar chart highlights that employee attrition is disproportionately concentrated in specific job roles, with Laboratory Technicians (26.16%), Sales Executives (24.05%), and Research Scientists (19.83%) collectively accounting for the vast majority of total attritions, indicating a critical need for targeted retention strategies within these technical and sales-oriented positions, while other roles like Healthcare Representatives, Managers, and Research Directors experience significantly lower turnover.

```python


# Filter only attrited employees
df_attrited = df[df['Attrition'] == 'Yes']

# Count attritions by age
age_counts = df_attrited['Age'].value_counts().sort_index()

# Calculate percentages
total_attrition = len(df_attrited)
age_percent = (age_counts / total_attrition * 100).round(2)

# Create DataFrame for plotting
age_df = pd.DataFrame({
    'Age': age_counts.index,
    'Attrition Count': age_counts.values,
    'Attrition %': age_percent.values
})

# Plot
plt.figure(figsize=(14,6))
barplot = sns.barplot(
    x='Age',
    y='Attrition Count',
    data=age_df,
    palette='viridis'
)

# Annotate each bar with % rotated 45°
for idx, row in age_df.iterrows():
    barplot.text(
        idx, 
        row['Attrition Count'] + 0.5, 
        f"{row['Attrition %']}%", 
        ha='right',
        rotation=90,
        fontsize=9,
        color='black'
    )

plt.title("Attrition Count by Age with Percentage of Total Attrition", fontsize=14)
plt.xlabel("Age")
plt.ylabel("Number of Attritions")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

The bar chart vividly illustrates that employee attrition is most concentrated among individuals in their late twenties and early thirties, with the peak occurring at ages 29 and 31, each accounting for 7.59% of total attritions, indicating a significant challenge in retaining this demographic. In contrast, attrition is noticeably lower for employees in their early twenties and becomes consistently very low for those in their forties and fifties, suggesting that the company's retention efforts should primarily target understanding and addressing the factors contributing to the high turnover among its mid-career workforce.

```python
# Categorize ages
bins = [18, 25, 30, 35, 45, 55, 100]
labels = ['18-25', '26-30', '31-35', '36-45', '46-55', '56+']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=True)

# Filter attrited employees
df_attrited = df[df['Attrition'] == 'Yes']

# Group by age group
age_group_counts = df_attrited['Age Group'].value_counts().sort_index()
age_group_percent = (age_group_counts / len(df_attrited) * 100).round(2)

# Build dataframe
age_group_df = pd.DataFrame({
    'Age Group': age_group_counts.index,
    'Attrition Count': age_group_counts.values,
    'Attrition %': age_group_percent.values
})

# Plot
plt.figure(figsize=(10,6))
barplot = sns.barplot(
    x='Age Group',
    y='Attrition Count',
    data=age_group_df,
    palette='rocket'
)

# Annotate with percentages
for idx, row in age_group_df.iterrows():
    barplot.text(
        idx,
        row['Attrition Count'] + 0.5,
        f"{row['Attrition %']}%",
        ha='center',
        fontsize=10,
        rotation=45
    )

plt.title("Attrition by Age Group with % of Total Attrition", fontsize=14)
plt.xlabel("Age Group")
plt.ylabel("Attrition Count")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Filter for attrited employees
df_attrited = df[df['Attrition'] == 'Yes']

# Count by gender
gender_counts = df_attrited['Gender'].value_counts()
gender_percent = (gender_counts / gender_counts.sum() * 100).round(2)

# Labels with percentage
labels = [f'{gender} ({pct}%)' for gender, pct in zip(gender_counts.index, gender_percent)]

# Donut chart
plt.figure(figsize=(6,6))
colors = ['#36a2eb', '#ff6384']  # Blue for Male, Pink for Female (editable)
wedges, texts = plt.pie(gender_counts, labels=labels, startangle=90, colors=colors)

# Donut hole
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Attrition by Gender", fontsize=14)
plt.tight_layout()
plt.show()
```

The "Attrition by Gender" donut chart reveals a significant disparity in employee turnover, with male employees constituting 63.29% of all attritions compared to 36.71% for female employees, indicating that factors driving departures are disproportionately affecting the male workforce within the organization.

```python

# Filter attrited employees
df_attrited = df[df['Attrition'] == 'Yes']

# Count attrition by Marital Status
marital_counts = df_attrited['Marital Status'].value_counts()
marital_percent = (marital_counts / marital_counts.sum() * 100).round(2)

# Create labels with percentages
labels = [f'{status} ({pct}%)' for status, pct in zip(marital_counts.index, marital_percent)]

# Donut Chart
plt.figure(figsize=(7,7))
colors = ['#FF9999','#66B2FF','#99FF99']  # Custom colors for better distinction
wedges, texts = plt.pie(marital_counts, labels=labels, startangle=90, colors=colors)

# Donut hole for visual appeal
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Attrition by Marital Status", fontsize=14)
plt.tight_layout()
plt.show()
```

The "Attrition by Marital Status" donut chart indicates that over half of all employee attritions, specifically 50.63%, originate from single individuals, significantly outpacing married employees (35.44%) and divorced employees (13.92%), suggesting that single employees represent the most mobile and retention-challenging demographic for the organization.

```python

# Filter attrited employees
df_attrited = df[df['Attrition'] == 'Yes']

# Count attrition by Type of Salary
type_counts = df_attrited['Type Of Salary'].value_counts()
type_percent = (type_counts / type_counts.sum() * 100).round(2)

# Create labels with percentages
labels = [f'{stype} ({pct}%)' for stype, pct in zip(type_counts.index, type_percent)]

# Donut Chart
plt.figure(figsize=(7,7))
colors = ['#FFD700','#C0C0C0','#CD7F32']  # Gold, Silver, Bronze to reflect High, Medium, Low
wedges, texts = plt.pie(type_counts, labels=labels, startangle=90, colors=colors)

# Donut hole for aesthetics
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Attrition by Type of Salary", fontsize=14)
plt.tight_layout()
plt.show()
```

The "Attrition by Type of Salary" donut chart clearly indicates that employees in the low salary bracket constitute the largest proportion of attritions at 47.68%, followed by those in the medium bracket at 32.49%, while high-salary earners have the lowest attrition at 19.83%, strongly suggesting that compensation is a primary driver of employee turnover, particularly among the lower-paid workforce.

```python

# Filter attrited employees
df_attrited = df[df['Attrition'] == 'Yes']

# Job Satisfaction Mapping (including future support for '5')
satisfaction_map = {
    1: "Very Low",
    2: "Low",
    3: "Neutral",
    4: "High"
    
}

# Map satisfaction levels and drop NaNs (if any)
df_attrited['Satisfaction Label'] = df_attrited['Job Satisfaction'].map(satisfaction_map)
df_attrited = df_attrited.dropna(subset=['Satisfaction Label'])

# Group and calculate counts and percentages
satisfaction_counts = df_attrited['Satisfaction Label'].value_counts().reindex(satisfaction_map.values(), fill_value=0)
satisfaction_percent = (satisfaction_counts / satisfaction_counts.sum() * 100).round(2)

# Plot
plt.figure(figsize=(8,6))
sns.barplot(
    x=satisfaction_counts.index,
    y=satisfaction_counts.values,
    palette="coolwarm"
)

# Annotate bars
for idx, value in enumerate(satisfaction_counts.values):
    plt.text(idx, value + 0.5, f"{satisfaction_percent.values[idx]}%", ha='center', fontsize=10, rotation=0)

plt.title("Attrition Count by Job Satisfaction Level", fontsize=14)
plt.xlabel("Job Satisfaction")
plt.ylabel("Attrition Count")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

The bar chart, "Attrition Count by Job Satisfaction Level," reveals that the highest proportions of employee departures stem from those with "Neutral" (30.8%) and "Very Low" (27.85%) job satisfaction, confirming that dissatisfaction is a primary driver of turnover; however, a substantial 21.94% of attrition also occurs among employees reporting "High" satisfaction, indicating that factors beyond mere job contentment likely influence their decision to leave.

```python

# Filter attrited employees
df_attrited = df[df['Attrition'] == 'Yes']

# Count and percentage by Education Field
edu_field_counts = df_attrited['Education Field'].value_counts()
edu_field_percent = (edu_field_counts / edu_field_counts.sum() * 100).round(2)

# Sort for consistency
edu_field_counts = edu_field_counts.sort_values(ascending=False)
edu_field_percent = edu_field_percent[edu_field_counts.index]

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x=edu_field_counts.index, y=edu_field_counts.values, palette="Spectral")

# Annotate
for i, val in enumerate(edu_field_counts.values):
    plt.text(i, val + 0.5, f"{edu_field_percent.values[i]}%", ha='center', fontsize=10)

plt.title("Attrition by Education Field", fontsize=14)
plt.xlabel("Education Field")
plt.ylabel("Attrition Count")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

The "Attrition by Education Field" chart prominently displays that employees with backgrounds in Life Sciences (37.55%) and Medical fields (26.58%) collectively constitute the vast majority of total attritions, far outweighing those from Marketing, Technical Degree, "Other," and Human Resources fields, strongly indicating a significant and concentrated talent retention challenge within these specific scientific and healthcare-related educational segments of the workforce.

```python
# Filter only attrited employees
df_attrited = df[df['Attrition'] == 'Yes']

# Count attrition by department
dept_attrition_counts = df_attrited['Department'].value_counts()

# Calculate percentage of total attrition
total_attritions = len(df_attrited)
dept_attrition_pct = (dept_attrition_counts / total_attritions * 100).round(2)

# Combine into a dataframe for plotting
dept_attrition_df = pd.DataFrame({
    'Department': dept_attrition_counts.index,
    'Attrition Count': dept_attrition_counts.values,
    'Attrition %': dept_attrition_pct.values
})

# Plot
plt.figure(figsize=(10,6))
barplot = sns.barplot(
    x='Department', 
    y='Attrition Count', 
    data=dept_attrition_df,
    palette='coolwarm'
)

# Annotate bars with percentage labels
for i, row in dept_attrition_df.iterrows():
    barplot.text(
        i, 
        row['Attrition Count'] + 1, 
        f"{row['Attrition %']}%", 
        color='black', 
        ha='center',
        fontsize=11
    )

plt.title("Attrition Count by Department with Percentage of Total Attrition", fontsize=14)
plt.xlabel("Department")
plt.ylabel("Number of Attritions")
plt.ylim(0, dept_attrition_df['Attrition Count'].max() + 10)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

The "Attrition Count by Department" chart distinctly shows that the vast majority of employee departures are concentrated within the Research & Development (56.12%) and Sales (38.82%) departments, which together account for over 94% of all attritions, while Human Resources experiences significantly lower turnover, clearly indicating that retention efforts must primarily focus on addressing the unique challenges within these two critical, high-attrition departments.

## Overall Attrition Insights

* **Demographic Patterns:**
    * **Age:** Attrition peaks significantly among employees in their late twenties and early thirties (ages 26-35).
    * **Marital Status:** Single employees are the most prone to leaving, accounting for over half of all attritions.
    * **Gender:** Male employees constitute a disproportionately high 63.29% of all attritions.

* **Organizational Concentration:**
    * **Departments:** The vast majority of attrition (over 94%) is concentrated within the **Research & Development** and **Sales** departments.
    * **Job Roles:** Key roles driving attrition include **Laboratory Technicians, Sales Executives, and Research Scientists**.

* **Educational Background:**
    * Employees with **Life Sciences and Medical education backgrounds** together account for over 60% of total attrition, indicating a specific challenge in retaining talent from these fields, likely linked to the high attrition in R&D roles.

* **Compensation as a Factor:**
    * **Salary Level:** Nearly half of all attritions (47.68%) originate from employees in the **low salary bracket**, strongly suggesting compensation competitiveness as a major issue.

* **Job Satisfaction Nuances:**
    * While "Neutral" and "Very Low" job satisfaction levels contribute the most to attrition, a significant portion of leavers also report "High" satisfaction, implying that factors beyond immediate job contentment (e.g., career growth, external opportunities, overall work environment) are also critical drivers of turnover.

* **Summary:** The company's attrition challenge is largely centered around its younger, lower-paid, technically/sales-oriented male workforce, predominantly in the Research & Development and Sales functions, highlighting a need for targeted retention strategies addressing compensation, career development, and specific departmental/role-related issues.
