```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("../data/hr_employee_attrition.xlsx")
print("Rows:", len(df))
df.head()
```

```python
print(df["Attrition"].value_counts())
print(f"\nAttrition rate: {(df['Attrition'] == 'Yes').mean() * 100:.1f}%")
```

```python
dept = df.groupby("Department")["Attrition"].apply(lambda x: (x == "Yes").mean() * 100)
dept.sort_values(ascending=False).plot(kind="bar", figsize=(8, 4), color="steelblue")
plt.title("Attrition % by department")
plt.ylabel("%")
plt.tight_layout()
plt.show()
```

```python
roles = df.groupby("Job Role")["Attrition"].apply(lambda x: (x == "Yes").mean() * 100)
roles.sort_values(ascending=False).head(8).plot(kind="barh", figsize=(8, 5), color="coral")
plt.title("Attrition % by job role (top 8)")
plt.xlabel("%")
plt.tight_layout()
plt.show()
```

```python
sat = df.groupby("Job Satisfaction")["Attrition"].apply(lambda x: (x == "Yes").mean() * 100)
sat.plot(kind="bar", figsize=(6, 4), color="seagreen")
plt.title("Attrition % by job satisfaction (1=low, 4=high)")
plt.ylabel("%")
plt.tight_layout()
plt.show()
```

```python
left = df[df["Attrition"] == "Yes"]
left["Age group"] = pd.cut(left["Age"], bins=[18, 25, 30, 35, 45, 55, 70], labels=["18-25", "26-30", "31-35", "36-45", "46-55", "56+"])
left["Age group"].value_counts().sort_index().plot(kind="bar", figsize=(7, 4), color="teal")
plt.title("Leavers by age group")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

**Summary:** ~16% attrition · Sales & Sales Rep highest · low satisfaction → more exits · most leavers in 20s–30s
